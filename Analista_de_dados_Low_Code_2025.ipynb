##Exemplo do for

```python
# Lista de pesos das pessoas (em kg)
pesos = [87, 35, 90, 70, 96, 67]

# Lista de alturas das pessoas (em metros)
alturas = [1.85, 1.41, 1.75, 1.70, 1.82, 1.74]

# Criação de uma lista vazia para armazenar os valores do IMC
imc = []

# Loop que percorre cada índice das listas de pesos e alturas
for i in range(len(pesos)):  # A função len(pesos) retorna o número de elementos na lista
    # Cálculo do IMC: peso dividido pelo quadrado da altura
    # round() é usado para arredondar o valor do IMC para 2 casas decimais
    imc.append(round(pesos[i] / alturas[i]**2, 2))  # Adiciona o IMC calculado à lista 'imc'

# Imprime a lista com os valores do IMC calculados
print(imc)
```

##Exemplo 4

```python
''''Imagine um sistema de controle de acesso para um evento, clube,
ou uma empresa onde os funcionários têm permissão para acessar
determinadas áreas. Quando a pessoa chega para entrar, o sistema solicita seu nome,
e ele precisa corresponder a um dos nomes registrados no sistema para que
o acesso seja autorizado. Se o nome não for encontrado, o sistema solicitará
o nome novamente até que o nome correto seja inserido.

Descrição do Cenário
O programa começa solicitando o nome da pessoa. Ele compara o
nome fornecido com os nomes registrados no sistema (armazenados em um dicionário de dados).
Se o nome fornecido corresponder a algum dos nomes no sistema, o acesso é permitido,
e o loop termina. Caso contrário, o sistema negará o acesso e solicitará novamente
o nome até que o nome correto seja fornecido.

Situação prática:
Crie um dicionario de nomes com nome e idade
Solicitação do nome: O sistema pede o nome da pessoa.
Verificação: O nome fornecido é comparado com a lista de nomes armazenados no sistema.
Acesso permitido: Se o nome corresponder, o sistema informa que o acesso foi permitido.
Acesso negado: Se o nome não corresponder, o sistema solicita o nome novamente
e só permite o acesso quando o nome correto for informado.'''

i= {
   1: {'nome':'joao','idade':20},
   2: {'nome':'maria','idade':30},
   3: {'nome':'jose','idade':40}
}

n = input('Qual o seu nome?').lower()
while True:
  for pessoa in i.values():
    if pessoa['nome'].lower() == n: # Se o nome for igual ao digitado ele vai permitir
      print('Acesso permitido')
      break
  else:
    print('Acesso Negado')
    n = input('Qual o seu nome?').lower() # Solicita o nome novamente
  break
```

```python
def somar_numeros(a, b):
    # Calcula a soma de dois números
    soma = a + b
    return soma

# Definindo dois números
numero1 = 8
numero2 = 5

# Chamando a função para calcular a soma
resultado = somar_numeros(numero1, numero2)

# Imprimindo o resultado
print(f"A soma de {numero1} e {numero2} é: {resultado}")
```

```python
def celsius_para_fahrenheit(celsius):
    # Converte Celsius para Fahrenheit
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Definindo a temperatura em Celsius
temperatura_celsius = 50

# Chamando a função para converter para Fahrenheit
temperatura_fahrenheit = celsius_para_fahrenheit(temperatura_celsius)

# Imprimindo o resultado
print(f"{temperatura_celsius}°C é igual a {temperatura_fahrenheit}°F")
```
