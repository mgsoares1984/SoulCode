```python
'''

Exercício:
Crie um dataframe com as seguintes medidas descritivas a partir do
df original:
1 - contagem - (count)
2 - media - (mean)
3 - mediana - (criar linha nova)
4 - moda - (criar linha nova)
5 - variância - (criar linha nova)
6 - desvio padrão - (std)
7 - valor mínimo - (min)
8 - Q1 - (25%)
9 - Q2 - (50%)
10 -Q3 - (75%)
11 -Valor máximo (max)
12 -AIQ - (criar linha nova)
13 -LI - (criar linha nova)
14 -LS -(criar linha nova)

'''
```

```python
medidas = df.describe()
print(medidas)
```

```python
#Transformando a linha de Q2 em uma lista chamada mediana
mediana = list(medidas.loc['50%'])
print(mediana)
```

```python
medidas.loc['mediana'] = mediana
medidas
```

```python
#forçando o df original a ter moda
df.loc[4,'altura']= 1.64
df.loc[3,'peso'] = 64
display(df)
```

```python
moda = [df['idade'].mode().loc[0],df['altura'].mode().loc[0],df['peso'].mode().loc[0],df['imc'].mode().loc[0]]
print(moda)
```

```python
#Crie uma lista vazia de vazia de variancia tranformandoo std em uma lista chamada de desvio, depois vou iterar sobre a lista desvio appedando o resultado na lista variancia

variancia = []
desvio = list(medidas.loc['std'])
for i in range(len(desvio)):
  variancia.append(round(desvio[i]**2,6))
print(variancia)
```

```python
'''

Tranformando primeiro e terceiro quartis em lista, criando listas vazias para aiq, li e ls, iterando sobre q1 e q3 para appendar na lista aiq

'''
#AIQ é amplitude interquartil
#AIQ uma medida estatística que descreve a dispersão dos dados centrais de um conjunto
'''Quanto maior o valor da AIQ, mais dispersos estão os dados
A AIQ é uma medida de dispersão dos 50% centrais dos dados
 A AIQ pode ser usada para identificar outliers '''

#Outlier são dados que se diferenciam drasticamente dos demais

q1 = list(medidas.loc['25%'])
q3 = list(medidas.loc['75%'])
aiq = []
li = []
ls = []
for i in range(len(q1)):
  aiq.append(round(q3[i]-q1[i],6))
print(aiq)
```

```python
'''Iterar sobre a lista aiq para appendar na listas li e ls'''
for i in range(len(aiq)):
  li.append(round(q1[i] - (1.5*aiq[i]),6))
  ls.append(round(q3[i] + (1.5*aiq[i]),6))
```

```python
''' Inserindo as listas criadas no df medidas'''

medidas.loc['moda'] = moda
medidas.loc['variancia'] = variancia
medidas.loc['LI'] = li
medidas.loc['LS'] = ls
medidas.loc['AIQ'] = aiq
display(medidas)
```

```python
'''
Renomeando as listas de acordo com os nomes pedidos no enunciado
'''
medidas.rename({'count':'contagem','mean':'média',
                'std':'desvio padrão','min':'valor mínimo','25%':'Q1',
                '50%':'Q2','75%':'Q3','max':'valor máximo'}, axis=0, inplace=True)

display(medidas)
```

```python
'''
Reordenando as linhas de acordo com o que foi solicitado no enunciado
'''
medidas = medidas.loc[['Contagem','Média','mediana','moda','variancia','Desvio Padrão', 'Valor Mínimo', 'Q1', 'Q2', 'Q3','Valor Máximo', 'AIQ - Amplitude Interquartil', 'LI- Limite Inferior', 'LS- Limite Superior']]
```

```python
''' Crie uma visão do salario médio por cidade'''

df[['cidade','salario']].groupby('cidade').mean().round(2).reset_index()
```

```python
'''crie uma visão do salario médio agrupado por departamento e por cargo'''

df[['departamento','cargo','salario']].groupby(['departamento','cargo']).mean().reset_index()
```

```python
'''
Crie uma visão dos 25% menores salarios agrupados por cargo
'''

df[['cargo','salario']].groupby('cargo').quantile(0.50).reset_index()
```

```python
'''
Vamos fazer uma consulta que retorne apenas os nomes e salarios cujo o salario seja maior que 20.000
'''

df.loc[df['salario'] > 20000,['nome','salario']]
```
