```python
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/aula-dados/dados-brutos/dados.csv')
display(df)
```

```python
df['Altura'] = (df['Altura']/100).round(2)
display(df)
```

```python
df['imc'] = (df['Peso']/df['Altura']**2).round(2)
display(df)
```

```python
df.to_csv('tratados.csv',index = False)
```
